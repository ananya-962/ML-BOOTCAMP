```python
ratings["total no of ratings"]=pd.DataFrame(df.groupby("movie").count()['ratings'])
```

# creating recommender system

```python
moviemat=df.pivot_table(index="user_id",columns="movie",values="ratings")
```

```python
moviemat.head()
```

```python
moviemat.shape
```

```python
starwars_user_ratings=moviemat['Star Wars (1977)']
```

```python
starwars_user_ratings.head(10)
```

```python
similar_to_starwars=moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_starwars
```

```python
corr_starwars=pd.DataFrame(similar_to_starwars,columns=['correlation'])
```

```python
corr_starwars=corr_starwars.join(ratings['total no of ratings'])
```

```python
corr_starwars
```

```python
corr_starwars[corr_starwars['total no of ratings']>100].sort_values('correlation',ascending=False)
```

```python
def predict_movies(movie_name):
    movie_user_ratings=moviemat[movie_name]
    similar_to_movie=moviemat.corrwith(movie_user_ratings)
    corr_movie=pd.DataFrame(similar_to_movie,columns=['correlation'])
    corr_movie.dropna
    corr_movie=corr_movie.join(ratings['total no of ratings'])
    predictions=corr_movie[corr_movie['total no of ratings']>250].sort_values('correlation',ascending=False)
    return predictions
```

```python
predict_my_movie=predict_movies("Titanic (1997)")
```
